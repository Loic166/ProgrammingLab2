```python
import dash
from dash import dcc, html, Input, Output
import numpy as np
from scipy.optimize import curve_fit
import scipy as sp
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns


x = np.linspace(0, 5, 80)
y = 2.5 * x + 1.2 + np.random.normal(0, 0.5, 80)

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Dashbord"),
    dcc.Slider(
        id = "id-slider",
        min = 1,
        max = 20,
        value = 1,
        step = 1,
        marks = {i: str(i) for i in range(1, 21)},
    ),
    dcc.Graph(id="id-val")
])

@app.callback(
    Output("id-val", "figure"),
    Input("id-slider", "value")
)

def fit_retta(grado):
    poly = PolynomialFeatures(degree=grado)
    x_poly = poly.fit_transform(x)

    modello = LinearRegression()
    modello.fit(x_poly, y)

    y_pred = modello.predict(x_poly)

    g = sns.scatterplot(x=x, y=y, name="X vs Y")
    g.add_t
    plt.show()

    return g


if __name__ == "__main__":
    app.run(debug=True)
```

```python
import dash
from dash import dcc, html, Input, Output
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Données synthétiques
np.random.seed(1)
X = np.linspace(-3, 3, 100)
y = 0.5 * X**3 - X**2 + X + 3 + np.random.randn(100) * 3
X = X.reshape(-1, 1)

# App Dash
app = dash.Dash(__name__)
app.title = "Régression polynomiale interactive"

# Layout
app.layout = html.Div([
    html.H2("Régression polynomiale dynamique", style={"textAlign": "center"}),

    dcc.Graph(id="graph", config={"displayModeBar": False}),

    html.Div([
        html.Label("Choisis le degré du polynôme :"),
        dcc.Slider(
            id="slider-degre",
            min=1,
            max=10,
            step=1,
            value=3,
            marks={i: str(i) for i in range(1, 11)},
            tooltip={"placement": "bottom", "always_visible": True}
        ),
    ], style={"width": "80%", "margin": "auto", "paddingTop": "20px"})
])

# Callback
@app.callback(
    Output("graph", "figure"),
    Input("slider-degre", "value")
)
def update_graph(degree):
    # Modèle de régression polynomiale
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, y)
    y_pred = model.predict(X_poly)

    # Tracer les données originales
    fig = px.scatter(
        x=X.flatten(), y=y,
        labels={"x": "X", "y": "y"},
        title=f"Régression polynomiale (degré {degree})",
        opacity=0.7,
    )

    # Ajouter la courbe de prédiction (douce avec spline)
    fig.add_trace(go.Scatter(
        x=X.flatten(),
        y=y_pred,
        mode="lines",
        name=f"Modèle degré {degree}",
        line=dict(color="red", width=3),
    ))

    fig.update_layout(template="plotly_white")

    return fig

# Lancer le serveur
if __name__ == "__main__":
    app.run(debug=True)
```
