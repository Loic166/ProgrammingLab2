## 🎲 Exercice 3 - Simulation de dé

```python
def lancer_de(n):
    lance = np.random.choice([1, 2, 3, 4, 5, 6], size=n)
    compteur = [np.sum(i == lance) for i in range(1, 7)]
    return np.array(compteur)

test = lancer_de(20)
print("Résultat :", test)
```

✅ **Fonction correcte. On aurait pu simplifier avec `np.bincount`, mais ta version est très bien.**

## 🎲 Exercice 4 - Fréquences moyennes

```python
import matplotlib.pyplot as plt

taille_echan = np.linspace(10, 1000, 100, dtype=int)
simulation = np.array([lancer_de(i) for i in taille_echan])
frequence = np.array([arr/sum(arr)*100 for arr in simulation])

face_1 = np.array([arr[0] for arr in frequence])
face_3 = np.array([arr[2] for arr in frequence])
face_6 = np.array([arr[5] for arr in frequence])

plt.figure(figsize=(10,6))
plt.plot(taille_echan, face_1, label="face 1")
plt.plot(taille_echan, face_3, label="face 3")
plt.plot(taille_echan, face_6, label="face 6")
plt.legend()
plt.title("Évolution moyenne des fréquences pour les faces 1, 3 et 6")
plt.xlabel("Taille de l'échantillon (nombre de lancés)")
plt.ylabel("Fréquence en %")
plt.grid(True)
plt.show()
```

✅ **Code et tracé impeccables.**

## 🧮 Exercice 7 - Réduction conditionnelle

```python
depenses = np.array([3200, 2750, 4100, 3600, 2950, 5100, 4400])

reduction = np.where(depenses < 3000, depenses * 0.85, depenses)
tot_eco = depenses.sum() - reduction.sum()

print("Montant total économisé :", tot_eco)
print("Dépenses ajustées :", reduction, ", total dépense ajusté :", reduction.sum())
```

✅ **Très bonne logique, bien implémenté.**

## 🔁 Exercice 8 - Fonction avec seed

```python
def simuler_jeu(seed, n):
    np.random.seed(seed)
    simulation = np.random.choice([1,2,3,4,5,6], size=n)
    f = np.array([sum(i==simulation)/n*100 for i in range(1,7)])
    return f

test = simuler_jeu(0, 5000)
print("test =", test)
```

✅ **Tout est bon. Résultat reproductible grâce au seed, fonction claire.**
