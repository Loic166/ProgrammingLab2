## Q1 ‚Äì Statistiques simples
Un tableau contient les ventes (en unit√©s) de 12 mois :
```python
ventes = np.array([105, 98, 120, 130, 90, 115, 102, 125, 110, 100, 95, 108])
```
1. Calcule la moyenne et l‚Äô√©cart-type.
2. Identifie les mois o√π les ventes sont sup√©rieures √† la moyenne + 1 √©cart-type.

---

````python
import numpy as np

ventes = np.array([105, 98, 120, 130, 90, 115, 102, 125, 110, 100, 95, 108])

mois = np.array(["janvier","f√©vrier","mars","avril","mai","juin","juillet","ao√πt","septembre","octobre","novembre","d√©cembre"])
#1
moyenne = ventes.mean()
ecrt_type = ventes.std()
print("Moyenne = ",moyenne)
print("Ecart-type = ",ecrt_type)

#2
indice_mois = np.where(ventes > moyenne+ecrt_type)
print(f"Les mois o√π les ventes sont sup√©rieur √† la moyenne + 1 ecart-type sont les mois de {mois[indice_mois][0]},{mois[indice_mois][1]} et {mois[indice_mois][2]}.")
````


## Q3 ‚Äì Simulation de lancers
√âcris une fonction `simuler_lancers(seed, n)` qui :
- Lance un d√© `n` fois avec une graine `seed`
- Retourne la fr√©quence en % de chaque face (1 √† 6)

Exemple :
```python
simuler_lancers(0, 1000) ‚Üí array([16.1, 18.0, 17.2, 16.9, 15.5, 16.3])
```
---

````python
def simuler_lancers(seed, n):
    np.random.seed(seed)
    univers = np.random.choice([1,2,3,4,5,6], n)
    freq = np.array([np.sum(i == univers)/n*100 for i in [1,2,3,4,5,6]])
    
    return freq

test = simuler_lancers(0, 1000)
print(test)
````

## Q4 ‚Äì Analyse de performance

- Sinon, garde la valeur inchang√©e

Retourne le total avant et apr√®s l'ajustement.
