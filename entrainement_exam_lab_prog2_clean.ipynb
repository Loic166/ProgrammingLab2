### Exercice 1 : Cr√©ation et op√©rations de base
- Cr√©e un tableau NumPy `a` avec les valeurs `[4, 8, 15, 16, 23, 42]`
- Calcule la somme, la moyenne et l'√©cart-type
- Multiplie tous les √©l√©ments par 2
- S√©lectionne les √©l√©ments sup√©rieurs √† 20

```python
# Ton code ici
import numpy as np

a = np.array([4, 8, 15, 16, 23, 42])
print("Somme = ",a.sum())
print("Moyenn = ", a.mean())
print("Ecar_type = ", a.std())

#mutiplication des √©l√©ment par 2
a = a*2
print("array a ajust√©: ",a)
# S√©lection des √©l√©ment sup√©rieur √† 20
print("Element de a sup√©rieur √† 20: ",a[a > 20])
```

### Exercice 3 : Fonction de simulation de d√©
- √âcris une fonction `lancer_de(n)` qui simule `n` lancers de d√© (valeurs de 1 √† 6)
- Retourne un tableau contenant le nombre de fois que chaque face est sortie

```python
# Ton code ici
def lancer_de(n):
    lance = np.random.choice([1,2,3,4,5,6], size=n)
    compteur = [np.sum(i == lance) for i in range(1,7)]
    return np.array(compteur)

test = lancer_de(20)
test
```

### Exercice 4 : Exp√©rience r√©p√©t√©e
- Utilise `lancer_de()` pour simuler 100 lancers pour des tailles d‚Äô√©chantillons de 10 √† 1000
- Calcule les fr√©quences relatives en % pour chaque simulation
- Trace l‚Äô√©volution moyenne des fr√©quences pour les faces 1, 3, 6

```python
# Ton code ici
import matplotlib.pyplot as plt

taille_echan = np.linspace(10, 1000, 100, dtype=int)
simulation = np.array([lancer_de(i) for i in taille_echan])
frequence = np.array([arr/sum(arr)*100 for arr in simulation])

face_1 = np.array([arr[0] for arr in frequence])
face_3 = np.array([arr[2] for arr in frequence])
face_6 = np.array([arr[5] for arr in frequence])

plt.figure(figsize=(10,6))
plt.plot(taille_echan, face_1, label="face 1")
plt.plot(taille_echan, face_3, label="face 3")
plt.plot(taille_echan, face_6, label="face 6")
plt.legend()
plt.title("Evolution moyen des fr√©quence pour les faces 1,3 et 6")
plt.xlabel("Taille de l'echantillon(nombre de lanc√©s)")
plt.ylabel("Frequence en %")
plt.grid(True)
plt.show()
```

### Exercice 7 : R√©ductions conditionnelles
Soit `depenses = np.array([3200, 2750, 4100, 3600, 2950, 5100, 4400])`
- Applique une r√©duction de 15% aux montants < 3000
- Calcule le montant total √©conomis√©
- Affiche les nouvelles d√©penses et le total ajust√©

```python
# Ton code ici
depenses = np.array([3200, 2750, 4100, 3600, 2950, 5100, 4400])

reduction = np.where(depenses < 3000, depenses*(1-.15), depenses)
tot_eco = depenses.sum() - reduction.sum()
print("Montant total √©conomis√© : ",tot_eco)
print("D√©penses ajust√©es : ",reduction,",total d√©pense ajust√© : ",reduction.sum())
```

### Exercice 8 : Fonction compl√®te avec seed
- √âcris une fonction `simuler_jeu(seed, n)` qui :
    - fixe le seed √† `seed`
    - simule `n` lancers de d√©
    - retourne la fr√©quence de chaque face
- Teste la fonction pour `seed=0` et `n=5000`

```python
# Ton code ici
def simuler_jeu(seed, n):
    np.random.seed(seed)
    simulation = np.random.choice([1,2,3,4,5,6], size=n)
    f = np.array([sum(i==simulation)/n*100 for i in range(1,7)])
    return f

test = simuler_jeu(0, 5000)
print("test = ",test)
```
